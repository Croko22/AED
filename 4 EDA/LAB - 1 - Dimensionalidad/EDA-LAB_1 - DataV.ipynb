```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
def euclidean_histogram(file_name,d):
  file = open(file_name, "r")
  data = list(csv.reader(file, delimiter=","))
  data = np.array(data[0], dtype=float)

  print(f"Min value: {np.min(data)} | Max value: {np.max(data)}")
  avg = np.average(data)
  print(f"Average value of the list: {avg}")
  print(f"SD of the list: {np.std(data)}")

  plt.hist(data, bins=None, rwidth=0.9, color='#607c8e')
  plt.grid()
  plt.title(f'Histograma de distancias entre puntos de dimension {d}')
  plt.xlabel('Distancia')
  plt.ylabel('Recurrencia')
```

```python
euclidean_histogram("d10_distances.csv",10)
```

```python
euclidean_histogram("d50_distances.csv",50)
```
